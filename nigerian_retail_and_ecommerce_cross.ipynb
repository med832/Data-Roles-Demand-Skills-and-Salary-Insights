# 🇳🇬 Nigerian Retail & E-commerce: Cross-Channel Sales Analysis (2024)

## 📌 Executive Summary
This project analyzes the cross-channel shopping behavior of Nigerian consumers in 2024. It explores the synergy between major online marketplaces like **Jumia, Konga, and Jiji** and established offline retail outlets such as **Shoprite, Computer Village, and Slot**.

### 📊 Key Objectives:
1. **Identify Spending Patterns:** Compare online vs. offline spending across major Nigerian cities.
2. **Channel Preference Analysis:** Understand where consumers prefer to shop and why.
3. **Cross-Channel Synergy:** Measure the frequency and impact of customers using both online and offline channels.
4. **Market Share Insights:** Evaluate the dominance of key players in the Nigerian retail ecosystem.


## 🛠️ Data Overview
The dataset contains 50,000 transactions with the following key attributes:
- **Transaction ID & Customer ID:** Unique identifiers.
- **Channels:** Online (Jumia, Konga, Jiji, etc.) and Offline (Shoprite, Spar, Game, etc.).
- **Spending:** Total spend in Nigerian Naira (NGN).
- **Geographic Data:** Analysis across 15 major Nigerian cities (Lagos, Abuja, Kano, Ibadan, etc.).
- **Behavioral Metrics:** Preferred channel and cross-channel purchase frequency.

```python
monthly_revenue = (
    df
    .groupby(
        'month'
    )
    ['total_revenue']
    .sum()
)
```

```python
(monthly_revenue / 1e9).plot(
    kind='bar',
    figsize=(10,5)
)

plt.title(
'Monthly Revenue (Billion)'
)

plt.xlabel(
'Month'
)

plt.ylabel(
'Revenue'
)

plt.show()
```

```python
channel_revenue = (
    df
    .groupby(
        'preferred_channel'
    )
    ['total_revenue']
    .sum()
)
```

```python
channel_avg = (
    df
    .groupby(
        'preferred_channel'
    )
    ['total_revenue']
    .mean()
)
```

```python
channel_avg.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title(
'Average Revenue by Channel'
)

plt.show()
```

```python
purchase_revenue = (
    df
    .groupby(
        'cross_channel_purchases'
    )
    ['total_revenue']
    .mean()
)
```

```python
plt.figure(
    figsize=(10,5)
)
purchase_revenue.plot(
    marker='o'
)
plt.ylim(
300000,
400000
)
plt.title(
'Average Revenue by Purchases'
)
plt.show()
```

```python
city_revenue = (
    df
    .groupby(
        'city'
    )
    ['total_revenue']
    .mean()
    .sort_values(
        ascending=False
    )
)
```

## 📈 Market Insights & Recommendations (2024)
Based on the data and current market trends in Nigeria:

### 1. The Rise of Hybrid Consumers
- A significant portion of consumers (approx. 33%) now utilize **both** online and offline channels.
- **Recommendation:** Retailers should implement O2O (Online-to-Offline) strategies, such as "Buy Online, Pick Up In-Store" (BOPIS).

### 2. Marketplace Dominance
- **Jumia** remains the market leader in online traffic, but **Jiji** and **Konga** show strong competition in specific categories like electronics and used goods.
- **Offline Giants:** Locations like **Computer Village (Lagos)** and **Slot** continue to be the primary hubs for high-value tech purchases due to trust and immediate fulfillment.

### 3. Geographic Hotspots
- **Lagos, Abuja, and Port Harcourt** account for the highest spend volumes.
- **Emerging Cities:** Cities like **Ibadan** and **Kano** show rapid growth in online channel adoption.

### 4. Strategic Outlook
- The Nigerian e-commerce market is projected to reach **$33 Billion by 2026**.
- **Key Success Factor:** Integrating mobile payments (OPay, PalmPay) and improving last-mile delivery logistics.
